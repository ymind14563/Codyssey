## 1. 추천 시스템이란?

- 추천 시스템(Recommendation System)은 사용자에게 적합한 아이템(영화, 책, 제품 등)을 제안하는 데 사용되는 소프트웨어 시스템
- 사용자의 과거 행동이나 아이템의 속성을 기반으로 관련성 높은 결과를 제공함

## 2. 추천 시스템의 목표

추천 시스템의 주요 목표는 사용자 경험을 향상시키고, 비즈니스 성과를 증대시키는 것

### 1) 사용자 만족: 개인화된 경험 제공.

### 2) 참여도 증가: 사용자가 더 많은 콘텐츠를 소비하도록 유도.

### 3) 수익 증대: 관련 상품을 추천해 구매 전환율을 높임.

### 4) 검색 효율화: 대규모 데이터에서 사용자가 원하는 결과를 더 빠르게 제시.

## 3. 추천 시스템의 종류

### 1) 콘텐츠 기반 추천 (Content-Based Recommendation)
### 2) 협업 필터링 (Collaborative Filtering)
### 3) 하이브리드 추천 (Hybrid Recommendation)
### 4) 지식 기반 추천 (Knowledge-Based Recommendation)
### 5) 딥러닝 기반 추천 (Deep Learning-Based Recommendation)


---

### 1) 콘텐츠 기반 추천 (Content-Based Recommendation)

#### (1) 개념
- 사용자가 과거에 소비하거나 선호한 아이템의 속성을 분석하여, 유사한 속성을 가진 아이템을 추천하는 방식.
- 사용자 개별 데이터를 분석하므로, 개인화 추천에 적합.

#### (2) 작동 방식
- 아이템의 속성을 벡터화 (예: 영화의 경우 장르, 배우, 감독, 키워드 등을 활용)
- 사용자가 선호했던 아이템의 벡터와 유사도를 계산.
- 코사인 유사도(Cosine Similarity), 유클리드 거리 등 사용.
- 유사도가 높은 아이템을 추천.

#### (3) 장점
- 개인화 가능: 사용자의 명시적 선호(평점, 리뷰)를 반영.
- 새로운 사용자 대응: 사용자가 조금만 데이터를 제공하면 추천 가능.
- 데이터 의존성 감소: 사용자 간 데이터 비교가 필요하지 않음.

#### (4) 단점
- Cold Start 문제 (아이템): 새로 추가된 아이템에 데이터가 없으면 추천이 어려움.
- 추천 다양성 부족: 사용자가 좋아했던 아이템과 비슷한 아이템만 추천하므로, 새로운 아이템을 발견하기 어려움.
- 복잡한 데이터 처리: 텍스트, 이미지 등의 속성을 처리하려면 추가적인 NLP 또는 Computer Vision 기술이 필요.

### 2) 협업 필터링 (Collaborative Filtering)

#### (1) 개념
- 사용자 간의 행동 데이터를 활용하여 추천하는 방식.
- 특정 사용자가 좋아하는 아이템을 비슷한 취향을 가진 사용자가 선호할 가능성이 높다고 가정함.

#### (2) 종류
- 사용자 기반 협업 필터링: 사용자 간의 유사도를 기반으로 추천 (예: A 사용자가 좋아하는 아이템을 B 사용자도 좋아할 가능성이 크다면, A가 선호한 아이템을 B에게 추천)
- 아이템 기반 협업 필터링: 아이템 간의 유사도를 기반으로 추천 (예: "이 영화를 본 사람은 저 영화도 봤습니다.")

#### (3) 작동 방식
- 사용자-아이템 상호작용 데이터를 행렬로 표현(예: 행: 사용자, 열: 아이템, 값: 평점, 클릭 수, 구매 횟수 등)
- 유사도 계산: 사용자 간 또는 아이템 간의 유사도를 계산 (코사인 유사도, 피어슨 상관계수 등이 사용)
- 유사도가 높은 사용자 또는 아이템을 기반으로 추천.

#### (4) 장점
- 다양한 아이템 추천 가능: 콘텐츠 기반 추천보다 더 많은 종류의 아이템 추천 가능.
- 명시적 속성 필요 없음: 아이템의 속성을 알지 않아도 사용자의 행동 데이터만으로 추천 가능.

#### (5) 단점
- Cold Start 문제 (사용자): 데이터가 부족한 신규 사용자에 대한 추천이 어려움.
- 희소성 문제: 대규모 사용자-아이템 데이터에서 상호작용이 적을 경우 추천 정확도가 떨어짐.

### 3) 하이브리드 추천 (Hybrid Recommendation)

#### (1) 개념
- 콘텐츠 기반 추천과 협업 필터링을 결합하여 각 방법의 단점을 보완하는 방식.
- 사용자의 과거 선호 데이터와 다른 사용자의 행동 데이터를 모두 활용.

#### (2) 구현 방법
- Weighted Hybrid: 콘텐츠 기반 추천과 협업 필터링의 결과를 가중 평균하여 결합.
- Switching Hybrid: 데이터 상황에 따라 가장 적합한 방법을 선택하여 추천.
- Feature Augmentation: 콘텐츠 기반 추천의 결과를 협업 필터링의 입력으로 활용하거나 그 반대로 진행.

#### (3) 장점
- Cold Start 문제 완화: 사용자 또는 아이템 데이터가 부족한 상황에서도 추천 가능.
- 추천 성능 향상: 두 가지 방법의 장점을 결합하여 높은 정확도와 다양성 제공.

#### (4) 단점
- 구현 복잡도 증가: 두 가지 방법을 결합하므로 계산량과 설계 복잡도가 높아짐.

---

### 4) 지식 기반 추천 (Knowledge-Based Recommendation)

#### (1) 개념
- 사용자로부터 직접 받은 입력 데이터를 바탕으로 추천하는 방식.
- 명시적 규칙과 논리를 기반으로 추천을 생성.

#### (2) 작동 방식
- 사용자로부터 선호 데이터를 수집. (예: "가격은 10만 원 이하, 색상은 검정.")
- 명시된 규칙에 따라 아이템 추천. (예: 조건을 만족하는 아이템을 필터링.)

#### (3) 장점
- 명확한 추천 근거 제공: 사용자 입력을 기반으로 하므로 추천의 이유를 명확히 설명할 수 있음.
- Cold Start 문제 해결: 사용자 행동 데이터가 없어도 동작 가능.

#### (4) 단점
- 사용자 입력 의존: 입력 데이터가 부정확하거나 제한적이면 추천 성능 저하.
- 추천 다양성 제한: 사용자가 미처 생각하지 못한 옵션은 추천되지 않음.

---

### 5) 딥러닝 기반 추천 (Deep Learning-Based Recommendation)

#### (1) 개념
- 딥러닝 모델을 사용해 사용자 행동 패턴과 아이템 간 복잡한 관계를 학습.
- 텍스트, 이미지, 비디오 등 다양한 데이터를 활용 가능.

#### (2) 작동 방식
- Embedding 학습: 사용자와 아이템을 고차원 공간에 임베딩하여 유사성을 학습.
- Sequential 모델링: 사용자의 시간 순서에 따른 행동 데이터를 학습하여 다음 행동을 예측.
- 멀티모달 학습: 텍스트, 이미지, 메타데이터를 결합하여 학습.

#### (3) 장점
- 복잡한 데이터 처리 가능: 다양한 데이터 유형(텍스트, 이미지 등)을 처리하여 정교한 추천 제공.
- 개인화 강화: 사용자 행동의 미묘한 차이를 학습.

#### (4) 단점
- 자원 소모: 모델 훈련과 추론에 많은 컴퓨팅 자원이 필요.
- Cold Start 문제: 충분한 데이터가 없는 상황에서는 성능이 저하.
